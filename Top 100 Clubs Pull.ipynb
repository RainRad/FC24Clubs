```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from splinter import Browser
from sqlalchemy import create_engine
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
dataframes = []

# The unique tags for both generations (Old Gen = cg4, Current Gen = cg5)
generations =  ["common-gen4", "common-gen5"]

# Iterate through each generation's top 100 leaderboard
for generation in generations:
    url = f"https://proclubs.ea.com/api/fc/allTimeLeaderboard?platform={generation}"
    browser.visit(url)
    site = soup(browser.html, "html.parser")

    # Find the <pre> tag containing the JSON data
    pre_tag = site.find('pre')
    
    if pre_tag:
        # Extract the JSON data from the <pre> tag
        json_data = pre_tag.text
        
        # Convert JSON to a Python dictionary
        json_dict = json.loads(json_data)
        
        # Create a DataFrame from the dictionary
        df = pd.DataFrame(json_dict)
        
        # Append the DataFrame to the list
        dataframes.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
final_df = pd.concat(dataframes, ignore_index=True)
```

```python
pd.set_option('display.max_rows', None)
```

```python
browser.quit()
```

```python
final_df.to_csv("T100Leaderboard.csv")
```
